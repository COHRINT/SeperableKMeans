```python
dims = [1,2,4]
sn = [100,400,700,1000]
mn = [4,10,15]
fn = [5,15,25]
dists = ['symKLD','JSD','euclid','EMD','bhatt']
rows_list = []
for dim in dims:
    for s in sn:
        for m in mn:
            for f in fn:
                for dist in dists:
                    df_sample = grab_data(data=data2,dims=dim,sn=s,mn=m,fn=f,dist=dist)
                    avg_isd = df_sample['ISD'].mean()
#                     print(type(avg_isd))
                    if m*f < s:
                        avg_time = df_sample['time'].mean()
                        run_isd_frac = df_sample['ISD']/df_sample['runnalls_ISD']
                        run_time_frac= df_sample['time']/df_sample['runnalls_time']
                        avg_run_isd_frac = run_isd_frac.mean()
                        avg_run_time_frac = run_time_frac.mean()
                        row = {'dim': dim, 'start_num': s, 'dist': dist, 'mid_num': m, 'fin_num': f, 'avg_isd': avg_isd,
                              'avg_time': avg_time, 'avg_runnalls_isd_frac': avg_run_isd_frac, 'avg_runnalls_time_frac': avg_run_time_frac}
                        rows_list.append(row)
                    else:
                        continue
avg_sample_df = pd.DataFrame(rows_list,columns=['dim','start_num','dist','mid_num','fin_num','avg_isd','avg_time','avg_runnalls_isd_frac','avg_runnalls_time_frac'])
# avg_sample_df.head()

avg_isd = {1: {}, 2: {}, 4: {}}
avg_row_list = []
sn_avg_row_list = []
for dim in dims:
    for dist in dists:
        avg_isd = avg_sample_df[(avg_sample_df['dist']==dist) \
                                           & (avg_sample_df['dim']==dim)]['avg_isd'].mean()
        avg_time = avg_sample_df[(avg_sample_df['dist']==dist) \
                                           & (avg_sample_df['dim']==dim)]['avg_time'].mean()
        avg_runnalls_isd = avg_sample_df[(avg_sample_df['dist']==dist) \
                                           & (avg_sample_df['dim']==dim)]['avg_runnalls_isd_frac'].mean()
        avg_runnalls_time = avg_sample_df[(avg_sample_df['dist']==dist) \
                                           & (avg_sample_df['dim']==dim)]['avg_runnalls_time_frac'].mean()
        for s in sn:
            s_avg_isd = avg_sample_df[(avg_sample_df['dist']==dist) & (avg_sample_df['start_num']==s)\
                                           & (avg_sample_df['dim']==dim)]['avg_isd'].mean()
            s_avg_time = avg_sample_df[(avg_sample_df['dist']==dist) & (avg_sample_df['start_num']==s) \
                                           & (avg_sample_df['dim']==dim)]['avg_time'].mean()
            s_avg_runnalls_isd = avg_sample_df[(avg_sample_df['dist']==dist) & (avg_sample_df['start_num']==s)\
                                           & (avg_sample_df['dim']==dim)]['avg_runnalls_isd_frac'].mean()
            s_avg_runnalls_time = avg_sample_df[(avg_sample_df['dist']==dist) & (avg_sample_df['start_num']==s)\
                                           & (avg_sample_df['dim']==dim)]['avg_runnalls_time_frac'].mean()
            sn_row = {'dim': dim, 'dist': dist, 'start_num': s, 'avg_isd': s_avg_isd,
                    'avg_time': s_avg_time, 'avg_runnalls_isd_frac': s_avg_runnalls_isd, 
                     'avg_runnalls_time_frac': s_avg_runnalls_time}
            sn_avg_row_list.append(sn_row)
        row = {'dim': dim, 'dist': dist, 'avg_isd': avg_isd,
                    'avg_time': avg_time, 'avg_runnalls_isd_frac': avg_runnalls_isd, 
                     'avg_runnalls_time_frac': avg_runnalls_time}   
        avg_row_list.append(row)

avg_df = pd.DataFrame(avg_row_list,columns = ['dim','dist','avg_isd','avg_time','avg_runnalls_isd_frac','avg_runnalls_time_frac']) 
sn_avg_df = pd.DataFrame(sn_avg_row_list,columns= ['dim','dist','start_num','avg_isd','avg_time','avg_runnalls_isd_frac','avg_runnalls_time_frac'])
```

```python
# look for evidence of overcondensation in symmetric KLD
import json
symkld_df = grab_data(data=data2,dist='symKLD')
print(symkld_df.shape)
symkld_df.head()
```

```python
cnt = 0
rows = []
for idx, row in symkld_df.iterrows():
    test_mix = json.loads(row['test_mix'])
    result_mix = json.loads(row['result_mix'])
    if row['mid_num']*row['fin_num'] > len(result_mix['weights']):
        cnt += 1
#         print('{} \t {} \t {} \t {} \t {}'.format(row['mid_num'],row['fin_num'],
#                 row['mid_num']*row['fin_num'],len(result_mix['weights']),
#                 row['mid_num']*row['fin_num'] - len(result_mix['weights'])))
        row_add = {'dim': row['dim'], 'start': row['start_num'],'mid': row['mid_num'],'fin': row['fin_num'],
                'tot_mixands': row['mid_num']*row['fin_num'], 
                'num_mixands': len(result_mix['weights']),
                'diff': row['mid_num']*row['fin_num'] - len(result_mix['weights'])}
        rows.append(row_add)
overcond_df = pd.DataFrame(rows,columns=['dim','start','mid','fin','tot_mixands','num_mixands','diff'])
print(cnt)

euclid_df = grab_data(data=data2,dist='euclid')
cnt = 0
rows = []
for idx, row in euclid_df.iterrows():
    test_mix = json.loads(row['test_mix'])
    result_mix = json.loads(row['result_mix'])
    if row['mid_num']*row['fin_num'] > len(result_mix['weights']):
        cnt += 1
#         print('{} \t {} \t {} \t {} \t {}'.format(row['mid_num'],row['fin_num'],
#                 row['mid_num']*row['fin_num'],len(result_mix['weights']),
#                 row['mid_num']*row['fin_num'] - len(result_mix['weights'])))
        row_add = {'dim': row['dim'], 'start': row['start_num'],'mid': row['mid_num'],'fin': row['fin_num'],
                'tot_mixands': row['mid_num']*row['fin_num'], 
                'num_mixands': len(result_mix['weights']),
                'diff': row['mid_num']*row['fin_num'] - len(result_mix['weights'])}
        rows.append(row_add)
overcond_df_euclid = pd.DataFrame(rows,columns=['dim','start','mid','fin','tot_mixands','num_mixands','diff'])
print(cnt)
```
